1. K-Means聚类算法的基本思想是将n个样本划分成k个聚类,使得同一个聚类中的样本之间的相似度最大,不同聚类之间的相似度最小。其算法流程如下:
(1)随机选择k个样本点作为初始聚类中心;
(2)计算每个样本点与聚类中心的距离,将样本点归入距离最近的聚类; 
(3)计算各聚类的类中心,作为新一轮的聚类中心;
(4)重复步骤(2)-(3),直到聚类中心不再变化或达到预设迭代次数。

2. K-Means算法通过计算样本点与聚类中心的距离,将样本点归入距离最近的类中,并不断更新类中心。具体来说,聚类中心的计算方法是,对于给定的聚类,计算该聚类中所有样本的均值,以该均值作为该聚类新的聚类中心。

3. K-Means算法对初始聚类中心的选择会影响最终的聚类结果。如果初始聚类中心抽取不合理,比如初始中心过于接近,则会导致不同的类别结果。通常的做法是进行多次运行,并选择聚类效果最好的结果。也可以考虑采用聚类中心初始化的优化方法,如k-means++算法。

4. K-Means算法的问题包括:对噪声数据和异常值敏感;需要预先指定聚类数k;聚类结果可能会陷入局部优化;无法发现非凸形的聚类等。可以通过结合其他算法进行改进。

5. K-Means应用场景包括:客户分群、图像分割、数据压缩等。这些场景需要将数据样本进行分类,K-Means可以快速有效地完成聚类。

6. 评估指标有:聚类间相似度、聚类内不相似度;Davies-Bouldin指数、Calinski-Harabasz指数等。这些指标综合反映了聚类效果。

### 回归（有监督）

* 给定一个特征矩阵x, 预测其对应的目标向量值y, 如果 y 是连续的, 那么这就是回归问题

```python
x=np.arange(10)
y=np.arange(10)+np.random.randn(10)#生成伪数据
```

```python
plt.plot(x,y,"o")
```

```python
xr=x[:,np.newaxis]#由于sklearn需要以特征矩阵的形式来展现，因此把x变成列的形式
xr
#这里一维只是一个特例，一般特征矩阵都是多维
```

```python
from sklearn.linear_model import LinearRegression
s1=LinearRegression(fit_intercept=True)
s1.fit(xr,y)
```

```python
s1.coef_
#s1.intercept_
```

```python
y_predict=s1.predict(xr)
y_predict
```

```python
plt.plot(x,y,"o")
plt.plot(x,y_predict,"-")
```
